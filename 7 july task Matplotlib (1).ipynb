DIPLAYING PLOTS IN MATPLOTLIB

```python
%matplotlib inline
x1=np.linspace(0,10,50)
plt.plot(x1,np.sin(x1),'-')
plt.plot(x1,np.cos(x1),'--')
plt.show()
```

SPECIFY BOTH LISTS

```python
import numpy as np
x3=np.arange(0.0,6.0,0.01)
plt.plot(x3, [xi**2 for xi in x3], 'b-')
plt.show()
```

```python
x9 = np.arange(0, 4, 0.2)

y9 = np.exp(-x9)

e1 = 0.1 * np.abs(np.random.randn(len(y9)))

plt.errorbar(x9, y9, yerr = e1, fmt = '.-')

plt.show();
```

ADDING A LEGEND

```python

x15 = np.arange(1, 5)
fig, ax = plt.subplots()
ax.plot(x15,x15*1.5)
ax.plot(x15,x15*3.0)
ax.plot(x15,x15/3.0)

ax.legend(['Normal','Fast','Slow']);
plt.show()

```

```python
x15 = np.arange(1,5)
fig, ax = plt.subplots()
ax.plot(x15,x15*1.5, label='Normal')
ax.plot(x15, x15*3.0, label='Fast')
ax.plot(x15, x15/3.0, label='slow')
ax.legend();
```
